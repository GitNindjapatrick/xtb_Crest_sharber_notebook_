###  <a id='toc1_1_'></a>[Fonction pour Extraire les nergies HOMO, LUMO et GAP](#toc0_)

Extraction des nergies `HOMO`, `LUMO` et `GAP`  partir des calculs raliss avec `PySCF`.

```python
from pyscf.data import nist
import time

#conversion en eV
au2ev = nist.HARTREE2EV

def find_homo_lumo(myhf, au2ev):
    """Function that returns the HOMO and LUMO index and the HOMO energy in eV

    Args:
        mf_pyscf (pyscf object): pyscf meam-field object of the molecule to be evaluated.
    """
    # Index of HOMO and LUMO
    lumo_idx = myhf.mo_occ.tolist().index(0.)
    homo_idx = lumo_idx - 1

    # Calculate the HOMO Homo-LUMO
    E_HOMO = myhf.mo_energy[homo_idx]*au2ev
    E_LUMO = myhf.mo_energy[lumo_idx]*au2ev
    E_g = abs(E_HOMO - E_LUMO)

    return E_HOMO, E_LUMO, E_g
```

###  <a id='toc1_1_'></a>[Fonction pour le Calcul des nergies HOMO, LUMO et GAP](#toc0_)


Calcul des nergies `HOMO, LUMO et GAP` avec `PySCF` en utilisant la base `B3LYP/6-31G(2df,p)` via les mthodes `DFT et HF`.


```python

from pyscf import dft,scf,gto
def pyscf_calculation(mol_xtb_xyz, directory):
    
    mol_pyscf = gto.Mole(
        atom=mol_xtb_xyz,
        charge=0,
        spin=0,
        basis="6-31G(2df,p)",
        symmetry=False,
        unit='Angstrom'
    )
    mol_pyscf.build()
    
    # Hartree-Fock calculation
    myhf = scf.RHF(mol_pyscf).run()
    
    # DFT calculation with B3LYP functional
    mdf = dft.RKS(mol_pyscf, xc="B3LYP").run()

    Eks_homo, Eks_lumo, Eks_g = find_homo_lumo(myhf, au2ev)
    Eks1_homo, Eks1_lumo, Eks1_g = find_homo_lumo(mdf, au2ev)
    
    
    return Eks_homo, Eks_lumo, Eks_g, Eks1_homo, Eks1_lumo, Eks1_g
```


###  <a id='toc1_1_'></a>[Fonction pour la Gnration des Coordonnes 3D RDKit](#toc0_)

Gnration des coordonnes molculaires avec `RDKit`.

```python
def generate_3d_rdkit_conformation(smiles_key, smiles, working_dir,path_xyz):
    """
    Generate RDKit 3D optimized conformation (an initial geometry) of a molecule from SMILES
    and store the xyz coordinates in a file
    """

    if not path_xyz.exists():
        try:
            # Defines a molecule from its SMILES string
            mol_rdkit = Chem.MolFromSmiles(smiles)

            if mol_rdkit is None:
                print(f"Could not create molecule from SMILES: {smiles}")
                return None

            # Add explicit Hs
            mol_rdkit = Chem.AddHs(mol_rdkit)

            # Generates the initial 3D conformation of the molecule
            result = AllChem.EmbedMolecule(mol_rdkit)
            if result != 0:  # Check if embedding was successful
                print(f"Embedding failed for molecule: {smiles_key}")
                return None

            # Optimizes the 3D conformation of the molecule using MMFF - Merck Molecular Force Field
            try:
                AllChem.MMFFOptimizeMolecule(mol_rdkit, maxIters=200, mmffVariant="MMFF94s")
            except ValueError as e:
                print(f"Optimization failed for molecule {smiles_key}: {e}")
                return None

            # Canonicalize the orientation of the conformation
            Chem.rdMolTransforms.CanonicalizeMol(mol_rdkit, normalizeCovar=True, ignoreHs=False)

            # Convert RDKit molecule to XYZ format
            mol_xyz = Chem.MolToXYZBlock(mol_rdkit)

            with open(path_xyz, "w") as f:
                f.write(mol_xyz)

            return path_xyz 

        except Exception as e:
            print(f"An error occurred while processing molecule {smiles_key}: {e}")
            return None
```

###  <a id='toc1_1_'></a>[ Fonction pour le Calcul des RMSD](#toc0_)

Calcul des valeurs `RMSD (Root Mean Square Deviation)`.

```python
def run_rmsd(reference_xyz, comparison_xyz, path_rmsd):
    try:
        # Dbut du chronomtrage
        start_rmsd = time.time()

        # Excuter la commande calculate_rmsd
        rmsd_process = sp.run(['calculate_rmsd', str(reference_xyz), str(comparison_xyz)], 
                              stdout=sp.PIPE, stderr=sp.PIPE, text=True, check=True)
        rmsd_time = time.time() - start_rmsd

        # Sauvegarder la sortie RMSD dans le fichier
        with open(path_rmsd, 'w') as f:
            f.write(rmsd_process.stdout)

        # Lire la valeur RMSD depuis le fichier
        with open(path_rmsd, 'r') as f:
            rmsd_value = float(f.readline().strip())

        return rmsd_value

    except sp.CalledProcessError as e:
        log_error(reference_xyz, "calculate_rmsd", str(e))
```

## <a id='toc1_'></a>[Qu'est-ce que MAE (Mean Absolute Error) ?](#toc0_)

### L'Erreur Absolue Moyenne (MAE) est une mesure statistique utilise pour quantifier la prcision des prdictions d'un modle en fonction des valeurs relles observes. Elle reprsente la moyenne des erreurs absolues entre les valeurs prdites et les valeurs observes, indpendamment du signe de l'erreur. L'quation pour calculer le MAE est la suivante :

$$
\mathtt{MAE} = \frac{1}{N} \sum_{i=1}^{N} |P_i - O_i|
$$

o :
- \( P_i \) reprsente les valeurs prdites,
- \( O_i \) reprsente les valeurs observes,
- \( N \) est le nombre total de points de donnes.
###  <h1 style="color: #cd7f32;" align="below">  **Exemples d'interprtation des valeurs de MAE :** </h1>



- **MAE faible (< 0.1)** : Les prdictions sont trs proches des valeurs relles, ce qui signifie une grande prcision.
- **MAE moyen (0.1 - 0.5)** : Les prdictions sont modrment proches des valeurs observes, mais avec des erreurs notables.
- **MAE lev (0.5 - 1.0)** : Il existe des carts substantiels entre les valeurs prdites et relles, indiquant une diminution de la prcision.
- **MAE trs lev (> 1.0)** : Les prdictions sont largement incorrectes par rapport aux valeurs relles.
---

Toutefois, son interprtation doit tre contextualise en fonction du domaine d'application et des caractristiques spcifiques des donnes.</strong>

## |<a id='toc1_'></a>[Qu'est-ce que le calibrage en rgression linaire ?](#toc0_)

Le calibrage est un processus qui permet d'ajuster un modle prdictif afin qu'il produise des prvisions qui soient aussi proches que possible des valeurs relles. En rgression linaire, cela implique souvent d'ajuster les coefficients du modle pour minimiser l'erreur entre les prdictions et les valeurs observes.

Formule Mathmatique

La rgression linaire peut tre exprime par l'quation suivante :

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n + \epsilon
$$

o :
- y  est la variable dpendante (valeur  prdire).
- $\beta_0 $ est l'ordonne  l'origine (intercept).
- $ \beta_1, \beta_2, ..., \beta_n $ sont les coefficients de rgression (pentes) pour chaque variable indpendante $ x_1, x_2, ..., x_n $.
- $ \epsilon $ est l'erreur alatoire.


##  <h1 style="color: #cd7f32;" align="below">  Calibrage du Modle </h1>


Le calibrage consiste  estimer les coefficients $beta_i$ de manire  minimiser l'erreur entre les valeurs prdites et les valeurs relles. Cela se fait gnralement en utilisant la mthode des moindres carrs, qui cherche  minimiser la somme des carrs des erreurs :

$$
\text{Minimiser} \quad S = \sum_{i=1}^{m} (y_i - \hat{y}_i)^2
$$

o :
- $ m$  est le nombre d'observations.
- $y_i$ est la valeur relle.
- $\hat{y}_i$ est la valeur prdite par le modle.

## <a id='toc1_'></a>[Qu'est-ce que RMSD(Root Mean Square Deviation) ?](#toc0_)


### L' cart quadratique moyen (RMSD) de deux structures est une mesure de la distance moyenne entre les atomes de deux structures superposes comme exemples  les coordonnes xyz de deux molcules. L'quation de base pour calculer le RMSD est la suivante :

$$ 
\mathtt{RMSD} = \sqrt{\frac{\sum_{i}^{N}{(O_i - P_i)}^2}{N}}
$$

o  $P_i$ et $Q_i$ sont les coordonnes des points correspondants dans les ensembles \( P \) et \( Q \), et \( N \) est le nombre total de points.

Q exemples d'interprtation des valeurs de RMSD :

- RMSD faible (< 0.5 ) : les deux molcules sont trs similaires, avec des diffrences mineures.
- RMSD moyen (0.5-1.5 ) : les deux molcules ont des diffrences notables, mais conservent une structure globale similaire.
- RMSD lev (1.5-3 ) : les deux molcules ont des diffrences importantes, avec des changements significatifs dans la structure.
- RMSD trs lev (> 3 ) : les deux molcules sont trs diffrentes, avec des structures quasi distinctes.

Cependant, il est important de noter que l'interprtation du RMSD dpend du contexte et de la nature des donnes. Il est recommand de consulter des experts dans le domaine pour obtenir une interprtation prcise et pertinente
