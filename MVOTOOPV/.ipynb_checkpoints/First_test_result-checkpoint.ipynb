## Introduction

Organic solar cells (OSCs) have emerged as promising candidates for next-generation renewable energy sources due to their flexibility, lightweight nature, and potential for low-cost production [1]. The efficiency of OSCs largely depends on the donor-acceptor compatibility, which influences charge transfer dynamics and electronic interactions. In this study, we assess the synthetic compatibility of GDB-9 molecules with two well-known materials: [6,6]-Phenyl-C61-butyric acid methyl ester (PCBM) and Poly[N-9'-heptadecanyl-2,7-carbazole-alt-5,5-(4',7'-di-2-thienyl-2',1',3'-benzothiadiazole) (PCDTBT) [2].

Computational chemistry provides a robust framework for predicting molecular interactions, electronic properties, and thermodynamic stability. Using Density Functional Theory (DFT) and Molecular Dynamics (MD) simulations, we explore the potential of GDB-9 molecules as donors in the active layer of OSCs.

## Computational Methods

To evaluate the synthetic compatibility of GDB-9 molecules with PCBM and PCDTBT, we employ the following computational approaches:

### 1. Density Functional Theory (DFT) Calculations

- **Software**: PySCF [3]  
- **Functional**: B3LYP  
- **Basis Set**: 6-31G(2df,p)  
- **Key Properties Calculated**:  
  - HOMO-LUMO gap  
  - Ionization potential (IP) and electron affinity (EA)  
  - Dipole moment  
  - Frontier molecular orbitals (FMOs)  

### 2. Molecular Dynamics (MD) Simulations

- **Software**: xTB CREST [4]  
- **Force Field**: GFN2-xTB  
- **Solvent**:  Toluene  
- **Analysis**:  
  - Radial distribution function (RDF)  
  - Interaction energy calculations  
  - Diffusion coefficients  

### 3. Binding Energy and Charge Transfer Analysis

- **Electronic Coupling Calculation**  
- **Non-covalent interaction (NCI) analysis**  
- **Charge transfer integral (J)**  

## Results and Discussion

### 1. Electronic Properties and Energy Level Alignment

- The HOMO and LUMO levels of GDB-9 molecules are analyzed to ensure favorable alignment with PCBM (acceptor) and PCDTBT (donor) [5].  
- Ionization potential and electron affinity values are used to assess charge transfer feasibility.  
- Molecular orbitals are visualized to examine charge delocalization and electronic interactions.  

### 2. Interaction Energies and Stability

- Binding energies between GDB-9/PCBM and GDB-9/PCDTBT pairs are computed using DFT calculations in PySCF.  
- Non-covalent interactions (π-π stacking, hydrogen bonding) are evaluated.  
- MD simulations with xTB CREST reveal the dynamic stability of these interactions in a solvent environment.  

### 3. Charge Transfer Efficiency

- The charge transfer integral (J) is computed to estimate electronic coupling.  
- Transition dipole moments and reorganization energies are analyzed to determine the efficiency of exciton dissociation.  
- The Scharber model is employed to predict photovoltaic efficiency [6].

## Z-score Calculation  

The **Z-score** is a statistical measure that indicates how much a data point deviates from the mean in terms of standard deviation. It is calculated using the following formula:  

$$
Z = \frac{X - \mu}{\sigma}
$$ 

where:  
- $X $ is the SAS score (or another metric) of a fragment,  
-  $\mu$ is the mean SAS score of all fragments,  
- $\sigma$ is the standard deviation of SAS scores.  

### Steps to Compute the Z-score from SMILES:  
1. Extract molecular fragments from SMILES.  
2. Calculate the **Synthetic Accessibility Score (SA Score)** for each fragment.  
3. Apply the Z-score formula to these values.  
4. Plot a **bar chart** with colors based on the Z-score, as shown in the figure.  


````python
import numpy as np
import pandas as pd

import seaborn as sns
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors, Descriptors
from scipy.stats import zscore


# Calcul du Z-score
df["Z_Score"] = zscore(df["sas1(%)"])

# Tri des fragments par Z-score pour un meilleur affichage
df = df.sort_values(by="Z_Score", ascending=True)

# Définition de la palette de couleurs (rouge pour négatif, vert pour positif)
colors = df["Z_Score"].apply(lambda x: "red" if x < 0 else "green")

# Tracé du graphique en barplot
plt.figure(figsize=(17, 15))
bars = plt.bar(df["smiles_key"], df["Z_Score"], color=colors, alpha=0.7)

# Ajout d'une ligne horizontale à Z = 0
plt.axhline(0, color='black', linestyle='dashed', linewidth=1)

# Ajout des étiquettes
plt.xlabel("Fragments (smiles_key)")
plt.ylabel("Z-Score")
plt.title("Barplot des Z-Scores des fragments moléculaires")

# Ajout d'une légende manuelle
plt.text(len(df) - 1, max(df["Z_Score"]), "Sur-représentés (verts)", color="green", ha="right")
plt.text(len(df) - 1, min(df["Z_Score"]), "Sous-représentés (rouges)", color="red", ha="right")

plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
````

````python
# Vérification des colonnes nécessaires
required_columns = {"pce_pcbm(%)", "sas1(%)", "smiles_key"}
if not required_columns.issubset(df.columns):
    missing_cols = required_columns - set(df.columns)
    raise ValueError(f"Colonnes manquantes dans le DataFrame : {missing_cols}")

# Filtrage des molécules avec PCE > 1
df_filtered = df[df["pce_pcbm(%)"] > 1].copy()

# Filtrage des outliers en utilisant les percentiles (5% - 95%)
q_low, q_high = df_filtered["pce_pcbm(%)"].quantile([0.05, 0.95])
df_filtered = df_filtered[(df_filtered["pce_pcbm(%)"] >= q_low) & (df_filtered["pce_pcbm(%)"] <= q_high)]

# Calcul des Z-scores pour identifier les fragments sur- et sous-représentés
df_filtered["Z_Score"] = zscore(df_filtered["pce_pcbm(%)"])

# Tri des fragments par Z-score pour une visualisation ordonnée
df_filtered = df_filtered.sort_values(by="Z_Score", ascending=True)

# Définition d'une palette de couleurs basée sur les Z-scores
colors = sns.color_palette("RdYlGn", as_cmap=True)
norm = plt.Normalize(df_filtered["Z_Score"].min(), df_filtered["Z_Score"].max())

# Création du barplot avec des couleurs indiquant la sur- ou sous-représentation
plt.figure(figsize=(20, 12))
bars = plt.bar(
    df_filtered["smiles_key"], 
    df_filtered["Z_Score"], 
    color=[colors(norm(z)) for z in df_filtered["Z_Score"]], 
    alpha=0.85
)

# Ajout d'une ligne horizontale pour indiquer Z = 0
plt.axhline(0, color='black', linestyle='dashed', linewidth=1)

# Rotation des labels pour une meilleure lisibilité
plt.xticks(rotation=90, ha="right", fontsize=10)

# Ajout des étiquettes pour les axes et du titre
plt.xlabel("Fragments (SMILES)", fontsize=14, labelpad=10)
plt.ylabel("Z-Score", fontsize=14, labelpad=10)
plt.title("Barplot des Z-Scores des molécules (PCE > 1) avec correction des outliers", fontsize=16, pad=15)

# Ajout d'annotations pour les fragments avec des Z-scores extrêmes
for i, z in enumerate(df_filtered["Z_Score"]):
    if z > 2 or z < -2:  # Seulement pour les valeurs Z > 2 ou Z < -2
        plt.text(i, z, f"{z:.2f}", fontsize=9, ha="center", color="black", fontweight="bold")

# Ajout d'une barre de couleur pour représenter les valeurs Z
sm = plt.cm.ScalarMappable(cmap=colors, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, pad=0.02)
cbar.set_label("Z-Score", fontsize=14)

# Ajustement des marges pour éviter l'encombrement
plt.tight_layout()

# Affichage du graphique
plt.show()
````

# Analyse des smiles dans les données GDB-9

## Introduction

Dans le cadre de la conception de matériaux pour le photovoltaïque organique (OPV), les smiles  moléculaires jouent un rôle essentiel dans l'optimisation des propriétés électroniques, telles que l'écart HOMO-LUMO, le facteur de remplissage (FF), et l'efficacité de conversion de puissance (PCE). Une analyse statistique basée sur les Z-scores permet d'identifier les smiles surreprésentés (positifs) et sous-représentés (négatifs) parmi les candidats ayant des performances élevées en termes de PCE.

Dans cette étude, nous utilisons les données de la base de données GDB-9 pour analyser les smiles  associés à  un **PCE > 1%**, en mettant en évidence leurs Z-scores. Cette approche est similaire à celle utilisée dans l'article de référence :  
**"Principes de conception et meilleurs candidats accepteurs non fullerènes pour le photovoltaïque organique"** ([source](https://www.sciencedirect.com/science/article/pii/S2542435117301307)).

---

## Méthodologie

1. **Filtrage des données :**
   - Seules les molécules avec un **PCE > 1%** sont sélectionnées.
   - Les fragments moléculaires associés à chaque SMILES sont extraits et analysés.

2. **Élimination des valeurs extrêmes :**
   - Les valeurs extrêmes sont corrigées en appliquant une plage de percentiles (5% - 95%).

3. **Calcul des Z-scores :**
   - Les Z-scores sont calculés pour la colonne `pce_pcbm(%)` (Synthetic Accessibility Score), permettant de normaliser et de comparer les fragments.

4. **Visualisation :**
   - Les Z-scores positifs (surreprésentés) sont affichés en **vert**.
   - Les Z-scores négatifs (sous-représentés) sont affichés en **rouge**.

---



## Discussion

Les résultats montrent une répartition claire des fragments moléculaires en fonction de leurs Z-scores :

1. **Fragments surreprésentés (Z-scores positifs) :**
   - Les fragments en vert, tels que `gdb 37079`, `gdb 37080`, `gdb 4847`, et `gdb 32866`, sont surreprésentés parmi les molécules avec un **PCE > 1%**.
   - Ces fragments sont généralement associés à des groupes aromatiques ou à des systèmes π-étendus, qui contribuent à **réduire l'écart HOMO-LUMO**, augmentant ainsi les valeurs de **Jsc** et de **PCE**.

2. **Fragments sous-représentés (Z-scores négatifs) :**
   - Les fragments en rouge, tels que `gdb 117032`, `gdb 326` et `gdb 37081`, sont sous-représentés parmi les molécules ayant un **PCE > 1%**.
   - Ces fragments incluent des groupes volumineux ou fortement électrophiles, susceptibles de provoquer une torsion le long du squelette moléculaire, **réduisant la conjugaison π** et diminuant les performances des matériaux.

3. **Corrélation avec la littérature :**
   - Les fragments surreprésentés identifiés dans cette étude concordent avec les résultats de l'article de référence, qui souligne l'importance des groupes aryles faiblement donneurs d'électrons pour optimiser les propriétés électroniques.
   - Les fragments sous-représentés, tels que les groupes terminaux volumineux (ex. indoles), sont également signalés dans la littérature pour leurs effets négatifs sur la conjugaison π et la planéité moléculaire.

---

## Conclusion

L'utilisation des Z-scores dans l'analyse des fragments moléculaires offre une méthode robuste pour identifier les sous-structures clés dans l'optimisation des matériaux OPV. Les principaux enseignements de cette analyse sont :

1. **Fragments bénéfiques :**
   - Les fragments conjugués en π, déficients en électrons, favorisent des écarts HOMO-LUMO réduits et des performances améliorées en termes de PCE.

2. **Fragments défavorables :**
   - Les fragments volumineux ou fortement électrophiles perturbent la planéité moléculaire et diminuent les performances.

Ces résultats confirment les principes de conception avancés dans l'article de référence, contribuant à guider le développement de matériaux OPV innovants.

Les résultats montrent que les molécules issues de Tartarus présentent une meilleure accessibilité synthétique que celles de GDB-9. Cette différence s'explique par les méthodologies de génération des bases de données : tandis que GDB-9 explore un large espace chimique sans se soucier de la faisabilité synthétique, Tartarus adopte une approche plus pragmatique, intégrant des contraintes de conception réalistes adaptées aux applications OPV.
````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.dummy import DummyRegressor
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler

# Fonction pour calculer 'Layer_Thickness' (exemple, remplacez par votre formule)
def calculate_layer_thickness(df):
    # Exemple de calcul : utiliser une combinaison d'autres colonnes
    # Assurez-vous que cela a du sens pour vos données
    df['Layer_Thickness'] = (df['jsc_pcbm(A.m-2)'] * df['voc_pcbm(V)']) / (df['pce_pcbm(%)'] + 1e-6) # Ajout d'une petite valeur pour éviter la division par zéro
    return df

# Appliquer le calcul de 'Layer_Thickness'
df = calculate_layer_thickness(df)


# Fonction pour transformer les SMILES en empreintes moléculaires
def smiles_to_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return np.zeros(2048)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
    arr = np.zeros((1,))
    DataStructs.ConvertToNumpyArray(fp, arr)
    return arr

# Appliquer la featurisation
df['Fingerprint'] = df['smiles'].apply(smiles_to_fingerprint)
X = np.vstack(df['Fingerprint'])

# Normalisation des caractéristiques (important pour kNN et potentiellement RF)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Utiliser les données normalisées

y_labels = ['pce_pcbm(%)', 'voc_pcbm(V)', 'jsc_pcbm(A.m-2)', 'Layer_Thickness']

# Paramètres pour GridSearchCV
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [4, 6, 8, 10],
    'min_samples_split': [2, 4],
    'min_samples_leaf': [1, 2]
}

param_grid_knn = {
    'n_neighbors': [5, 10, 15, 20],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}


# Modèles et résultats
results = []
best_models = {}  # Dictionnaire pour stocker les meilleurs modèles
for label in y_labels:
    y = df[label].values
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=142) # Utiliser les données normalisées
    
    # Dummy regressor
    dummy = DummyRegressor(strategy='mean')
    dummy.fit(X_train, y_train)
    y_pred_dummy = dummy.predict(X_test)
    dummy_mae = mean_absolute_error(y_test, y_pred_dummy)
    dummy_r2 = r2_score(y_test, y_pred_dummy)
    
    # Random Forest avec GridSearchCV
    grid_search_rf = GridSearchCV(RandomForestRegressor(random_state=142), param_grid_rf, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1)
    grid_search_rf.fit(X_train, y_train)
    best_rf = grid_search_rf.best_estimator_
    y_pred_rf = best_rf.predict(X_test)
    rf_mae = mean_absolute_error(y_test, y_pred_rf)
    rf_r2 = r2_score(y_test, y_pred_rf)
    best_models[f'RF_{label}'] = best_rf # Stocker le meilleur modèle
    
    # kNN avec GridSearchCV
    grid_search_knn = GridSearchCV(KNeighborsRegressor(), param_grid_knn, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1)
    grid_search_knn.fit(X_train, y_train)
    best_knn = grid_search_knn.best_estimator_
    y_pred_knn = best_knn.predict(X_test)
    knn_mae = mean_absolute_error(y_test, y_pred_knn)
    knn_r2 = r2_score(y_test, y_pred_knn)
    best_models[f'kNN_{label}'] = best_knn # Stocker le meilleur modèle
    
    # Stocker les résultats
    results.append([label, dummy_mae, rf_mae, knn_mae, dummy_r2, rf_r2, knn_r2])

# Convertir en DataFrame
results_df = pd.DataFrame(results, columns=["Property", "Dummy MAE", "RF MAE", "kNN MAE", "Dummy R2", "RF R2", "kNN R2"])
print(results_df)

# Visualisation des prédictions
num_plots = len(y_labels)
fig, axes = plt.subplots(num_plots, 2, figsize=(12, 5 * num_plots))  # Ajuster la taille pour plus de clarté

for i, label in enumerate(y_labels):
    y_test = df.iloc[X_train.shape[0]:][label].values  # Sélectionner les valeurs de test correctes
    
    # Prédictions avec les meilleurs modèles
    y_pred_knn = best_models[f'kNN_{label}'].predict(X_test)
    y_pred_rf = best_models[f'RF_{label}'].predict(X_test)
    
    # kNN Plot
    ax_knn = axes[i, 0]
    sns.regplot(x=y_test, y=y_pred_knn, ax=ax_knn, scatter_kws={'alpha':0.5}, line_kws={'color': 'red'}) # Ajout d'une régression line
    ax_knn.set_xlabel(f'Ground Truth {label}')
    ax_knn.set_ylabel(f'Predicted {label} (kNN)')
    ax_knn.set_title(f'kNN: {label}')
    
    # Random Forest Plot
    ax_rf = axes[i, 1]
    sns.regplot(x=y_test, y=y_pred_rf, ax=ax_rf, scatter_kws={'alpha':0.5}, line_kws={'color': 'red'}) # Ajout d'une régression line
    ax_rf.set_xlabel(f'Ground Truth {label}')
    ax_rf.set_ylabel(f'Predicted {label} (RF)')
    ax_rf.set_title(f'Random Forest: {label}')

plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import numpy as np
import shap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
from mordred import Calculator, descriptors

calc = Calculator(descriptors, ignore_3D=True)

def calculate_layer_thickness(df):
    df['Layer_Thickness'] = (df['jsc_pcbm(A.m-2)'] * df['voc_pcbm(V)']) / (df['pce_pcbm(%)'] + 1e-6)
    return df

def smiles_to_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return np.zeros(2048)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
    arr = np.zeros((1,))
    DataStructs.ConvertToNumpyArray(fp, arr)
    return arr

def smiles_to_descriptors(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    desc = calc(mol)
    return pd.Series(desc).values

def clean_features(X):
    X = np.nan_to_num(X, nan=0.0, posinf=0.0, neginf=0.0)
    return X

def train_and_evaluate_models(df, y_labels):
    df = calculate_layer_thickness(df)
    df['Fingerprint'] = df['smiles'].apply(smiles_to_fingerprint)
    df['Descriptors'] = df['smiles'].apply(smiles_to_descriptors)
    df = df.dropna(subset=['Descriptors'])

    X_fp = np.vstack(df['Fingerprint'])
    X_desc = np.vstack(df['Descriptors'])
    X_desc = clean_features(X_desc)
    X = np.hstack((X_fp, X_desc))

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    results = []
    best_models = {}

    for label in y_labels:
        y = df[label].values
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=142)
        
        param_grid = {
            'n_estimators': [200, 300],
            'max_depth': [10, 20],
            'min_samples_split': [2, 5]
        }
        
        grid_search = GridSearchCV(RandomForestRegressor(random_state=142), param_grid, cv=3, n_jobs=-1)
        grid_search.fit(X_train, y_train)
        
        best_model = grid_search.best_estimator_
        y_pred = best_model.predict(X_test)
        
        best_models[f'RandomForest_{label}'] = best_model
        results.append({
            "Propriété": label,
            "MAE": mean_absolute_error(y_test, y_pred),
            "RMSE": mean_squared_error(y_test, y_pred, squared=False),
            "R2": r2_score(y_test, y_pred)
        })
    
    results_df = pd.DataFrame(results)
    print(results_df)
    
    return best_models, X_test, scaler


def explain_model(best_models, X_test, y_labels, scaler):
    # Vérifier que X_test est bien transformé avec le même scaler
    X_test_scaled = scaler.transform(X_test)

    for label in y_labels:
        model_key = f'RandomForest_{label}'
        if model_key in best_models:
            explainer = shap.Explainer(best_models[model_key], X_test_scaled)
            shap_values = explainer(X_test_scaled)  # Désactiver l'additivité si nécessaire
            shap.summary_plot(shap_values, X_test_scaled)
        else:
            print(f"Modèle pour {label} non trouvé.")

# Exemple d'utilisation :
y_labels = ['pce_pcbm(%)', 'voc_pcbm(V)', 'jsc_pcbm(A.m-2)', 'Layer_Thickness']

# Après chargement de votre dataframe :
best_models, X_test, scaler = train_and_evaluate_models(df, y_labels)

# Appeler la fonction explain_model avec le scaler inclus.
explain_model(best_models, X_test, y_labels, scaler)



````

````python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem
from mordred import Calculator, descriptors
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler


# ============================
# Étape 1 : Calcul des Descripteurs Mordred
# ============================


def get_mordred_dict(mol: Chem.Mol) -> dict:
    """
    Calcule les descripteurs Mordred pour une molécule donnée.

    Args:
        mol: Molécule RDKit

    Returns:
        Descripteurs Mordred sous forme de dictionnaire
    """
    calc = Calculator(descriptors, ignore_3D=True)
    descriptors_dict = calc(mol).asdict()
    return descriptors_dict

def generate_mordred_descriptors(donor_structures: pd.DataFrame) -> pd.DataFrame:
    """
    Génère les descripteurs Mordred pour les molécules en format smiles.

    Args:
        donor_structures: DataFrame contenant une colonne "smiles"

    Returns:
        DataFrame des descripteurs filtrés
    """
    
    donor_mols = donor_structures["smiles"].map(lambda smiles: Chem.MolFromSmiles(smiles))
    
    # Calcul des descripteurs
    print("Calcul des descripteurs Mordred...")
    descriptors = donor_mols.apply(get_mordred_dict)
    mordred_descriptors = pd.DataFrame.from_records(descriptors, index=donor_structures.index)
    
    # Nettoyage des données
    mordred_descriptors = mordred_descriptors.select_dtypes(exclude=["object"])  # Retirer les erreurs de calcul
    mordred_descriptors.dropna(axis=1, how='any', inplace=True)  # Supprimer les colonnes avec des NaN
    descriptor_variances = mordred_descriptors.var(numeric_only=True)  # Calculer la variance des colonnes
    invariant_descriptors = descriptor_variances[descriptor_variances == 0].index  # Colonnes invariantes
    mordred_descriptors.drop(columns=invariant_descriptors, inplace=True)
    
    print("Descripteurs Mordred générés et nettoyés.")
    return mordred_descriptors


# Calculer les descripteurs
mordred_descriptors = generate_mordred_descriptors(df)
mordred_descriptors.head()  # Affiche un aperçu des descripteurs générés
# ============================
# Étape 3 : PCA
# ============================
print("Standardisation et PCA...")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(mordred_descriptors)  # Utiliser mordred_descriptors
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
df[["PCA1", "PCA2", "PCA3"]] = X_pca  # Affecter directement

# ============================
# Étape 4 : Clustering (K-Means) et Analyse de Silhouette
# ============================
range_n_clusters = range(2, min(25, len(df) + 1)) #Ensure range is valid even for small datasets
silhouette_scores = []
cluster_labels_list = [] # Garder trace des labels de clusters

print("Détermination du nombre optimal de clusters...")
for n_clusters in range_n_clusters:
    print(f"  Essai avec n_clusters = {n_clusters}...") # Indicateur de progression
    kmeans = KMeans(n_clusters=n_clusters, random_state=142, n_init='auto')
    cluster_labels = kmeans.fit_predict(X_pca)
    silhouette_avg = silhouette_score(X_pca, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    cluster_labels_list.append(cluster_labels)
    print(f"    Silhouette score: {silhouette_avg:.3f}")

# Visualisation du score de silhouette
plt.figure(figsize=(8, 6))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title("Silhouette Score vs Nombre de Clusters")
plt.xlabel("Nombre de Clusters")
plt.ylabel("Silhouette Score")
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()

# Choix du nombre optimal de clusters
if silhouette_scores:  # Vérifier si silhouette_scores n'est pas vide
    optimal_n_clusters = range_n_clusters[np.argmax(silhouette_scores)]
    print(f"Nombre optimal de clusters: {optimal_n_clusters}")

    # Clustering final avec le nombre optimal
    kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=142, n_init='auto')
    df["Cluster"] = kmeans.fit_predict(X_pca)
    cluster_stats = df.groupby(['Cluster', 'voc_pcbm(V)']).agg(
        mean_pl=('jsc_pcbm(A.m-2)', 'mean'),
        max_pl=('jsc_pcbm(A.m-2)', 'max')
    ).reset_index()
else:
    print("Aucun score de silhouette calculé. Impossible de déterminer le nombre optimal de clusters.")
    optimal_n_clusters = None  # Définir optimal_n_clusters à None pour éviter des erreurs ultérieures

# ============================
# Étape 5 : Visualisations
# ============================
if optimal_n_clusters is not None:
    # Visualisation 2D des clusters
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df["PCA1"], y=df["PCA2"], hue=df["Cluster"], palette="viridis", s=60)
    plt.xlabel("PCA1")
    plt.ylabel("PCA2")
    plt.title("Clusters visualisés en 2D (PCA)")
    plt.show()

    # Visualisation 3D des clusters
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    scatter = ax.scatter(df["PCA1"], df["PCA2"], df["PCA3"], c=df["Cluster"], cmap="viridis", s=60)
    ax.set_xlabel("PCA1")
    ax.set_ylabel("PCA2")
    ax.set_zlabel("PCA3")
    plt.title("Clusters visualisés en 3D (PCA)")
    plt.show()

    # Performances par cluster (boxplot)
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df["Cluster"], y=df["pce_pcbm(%)"], palette="viridis")
    plt.xlabel("Cluster")
    plt.ylabel("pce_pcbm(%)")
    plt.title("Distribution des performances par cluster")
    plt.show()


else:
    print("Impossible de générer les visualisations car aucun cluster n'a été créé.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem import Draw

# ==============================================================================
# 1. DATA LOADING
# =============================================================================
# ==============================================================================
# 2. FRAGMENT EXTRACTION
# ==============================================================================

def extract_fragments(smiles, radius=2, nBits=2048):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        try:
            fingerprint = rdMolDescriptors.GetMorganFingerprintAsBitVect(mol, radius=radius, nBits=nBits)
            return list(fingerprint.GetOnBits())
        except:
            return []
    return []

fragment_count = {}
for smiles in df["smiles"]:
    fragments = extract_fragments(smiles)
    for bit in fragments:
        fragment_count[bit] = fragment_count.get(bit, 0) + 1

fragments_df = pd.DataFrame(list(fragment_count.items()), columns=["Fragment", "Frequency"])
fragments_df = fragments_df.sort_values(by="Frequency", ascending=False).reset_index(drop=True)

# ==============================================================================
# 3. VISUALIZATION
# ==============================================================================

plt.figure(figsize=(10, 7))
plt.plot(range(1, len(fragments_df) + 1), np.log10(fragments_df["Frequency"]), color='darkred', linewidth=1.5)
plt.xlabel("Fragment Rank (by Frequency)", fontsize=14)
plt.ylabel("Log10(Frequency)", fontsize=14)
plt.title("Distribution of Molecular Fragment Frequencies", fontsize=16, fontweight='bold')
plt.grid(True, linestyle="--", alpha=0.7)
plt.savefig("distribution_fragments.png", dpi=400)
plt.show()

# ==============================================================================
# 4. CORRECTION DE VISUALIZATION DES FRAGMENTS
# ==============================================================================

def visualize_fragments(fragment_bits, n_fragments=5, molsPerRow=5, subImgSize=(200, 200)):
    mols = []
    for bit in fragment_bits[:n_fragments]:
        dummy_mol = Chem.MolFromSmiles('C')  # Molécule de base
        info = {}

        # Générer l'empreinte avec bitInfo
        rdMolDescriptors.GetMorganFingerprint(dummy_mol, 2, bitInfo=info)

        # Extraire et dessiner le fragment
        if bit in info:
            for atomId, radius_list in info.items():  
                radius = radius_list[0]  # ✅ Correction : Prendre le premier élément (correct)
                env = Chem.FindAtomEnvironmentOfRadiusN(dummy_mol, radius, atomId)
                if not env:  # ✅ Vérification pour éviter l'erreur
                    continue
                atom_map = {}
                submol = Chem.PathToSubmol(dummy_mol, env, atomMap=atom_map)
                Chem.Kekulize(submol)
                mols.append(submol)

    if not mols:  # ✅ Vérification pour éviter un affichage vide
        print("Aucun fragment valide n'a été généré.")
        return

    img = Draw.MolsToGridImage(mols, molsPerRow=molsPerRow, subImgSize=subImgSize,
                               legends=[f"Fragment {i+1}" for i in range(len(mols))])

    # ✅ Correction de l'affichage avec PIL
    plt.imshow(np.array(img))
    plt.axis('off')
    plt.show()

# Tester la visualisation avec les fragments les plus fréquents
top_fragments = fragments_df["Fragment"].head(5).tolist()
visualize_fragments(top_fragments)
````

````python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import rdMolDescriptors
import matplotlib.pyplot as plt
import seaborn as sns

# ============================
# 1. Chargement des Données
# ============================

print("Colonnes disponibles :", df.columns)

# ============================
# 2. Définition des Blocs de Construction
# ============================

building_blocks = {
    1: "c1ccccc1",  # Benzène
    2: "C=C",       # Double liaison
    3: "c1ccc(cc1)N",  # Phénylamine
    4: "c1ccccc1O",  # Phénol
    5: "c1ccccc1S",  # Thiophène
    6: "c1ccccc1Se",  # Séléniophène
    7: "c1ccccc1F",  # Fluorobenzène
    8: "c1ccccc1Cl",  # Chlorobenzène
    9: "c1ccccc1Br",  # Bromobenzène
    10: "c1ccccc1I",  # Iodobenzène
    11: "c1ccncc1",   # Pyridine
    12: "c1cccnc1",   # Pyrrole
    13: "C#N",        # Cyano
    14: "C=O",        # Carbonyle
    15: "c1ccccc1C",  # Toluène
    16: "c1ccccc1N",  # Aniline
    17: "c1cnccn1",   # Imidazole
    18: "c1ncccc1",   # Pyrimidine
    19: "C1=CC=C(C=C1)C",  # Benzyle
    20: "c1ccscc1",   # Thiophène
    21: "C1=CN=CN1",  # Pyridazine
    22: "C1=CC=CN1",  # Isoquinoline
    23: "N",          # Azote
    24: "O",          # Oxygène
    25: "S",          # Soufre
    26: "Se"          # Sélénium
}

# ============================
# 3. Identification des Blocs dans les SMILES
# ============================

def count_blocks(smiles, building_blocks):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return {k: 0 for k in building_blocks.keys()}  # Si la molécule est invalide, retourner 0 pour tous les blocs
    block_counts = {}
    for idx, smarts in building_blocks.items():
        pattern = Chem.MolFromSmarts(smarts)
        if pattern:
            count = len(mol.GetSubstructMatches(pattern))  # Compter les occurrences du motif
        else:
            count = 0
        block_counts[idx] = count
    return block_counts

# Calculer les occurrences des blocs pour chaque molécule
print("Calcul des occurrences des blocs de construction...")
block_occurrences = df['smiles'].apply(lambda x: count_blocks(x, building_blocks))
block_df = pd.DataFrame(block_occurrences.tolist())

# Supprimer les colonnes dupliquées après concaténation
block_df.columns = [f"block_{i+1}" for i in range(block_df.shape[1])]

# Fusionner les données en supprimant les duplications de colonnes
df = pd.concat([df.reset_index(drop=True), block_df.reset_index(drop=True)], axis=1)

# ============================
# 4. Analyse des Occurrences
# ============================

threshold = df['pce_pcbm(%)'].quantile(0.9)  # Seuil pour les meilleurs candidats
top_candidates = df[df['pce_pcbm(%)'] >= threshold]
top_frequencies = top_candidates.iloc[:, -len(building_blocks):].mean()  # Moyenne des occurrences par bloc

overall_frequencies = df.iloc[:, -len(building_blocks):].mean()  # Moyenne des occurrences par bloc

relative_frequencies = (top_frequencies - overall_frequencies) / overall_frequencies

# ============================
# 5. Visualisation des Résultats
# ============================

visualization_data = pd.DataFrame({
    "Block": list(building_blocks.keys()),
    "Relative Frequency": relative_frequencies,
    "Label": ["Amplified" if x > 0 else "Decreased" for x in relative_frequencies]
})

# Tracer les blocs amplifiés et diminués
plt.figure(figsize=(10, 6))
sns.barplot(
    data=visualization_data,
    x="Block",
    y="Relative Frequency",
    hue="Label",
    palette={"Amplified": "green", "Decreased": "red"}
)
plt.axhline(0, color="black", linewidth=1, linestyle="--")
plt.title("Relative Frequencies of Building Blocks")
plt.xlabel("Building Block ID")
plt.ylabel("Relative Frequency")
plt.xticks(rotation=90)
plt.legend(title="Occurrence", loc="upper right")
plt.tight_layout()
plt.show()

# ============================
# 6. Extraction des SMILES Amplifiés et Diminués
# ============================

# Extraction des SMILES pour les blocs les plus amplifiés et diminués
max_amplified_block = visualization_data.loc[visualization_data["Relative Frequency"].idxmax(), "Block"]
max_decreased_block = visualization_data.loc[visualization_data["Relative Frequency"].idxmin(), "Block"]

# Convertir les indices de blocs pour correspondre aux colonnes concaténées
max_amplified_col = f"block_{max_amplified_block}"
max_decreased_col = f"block_{max_decreased_block}"

# SMILES des molécules contenant le bloc le plus amplifié
amplified_smiles = df[df[max_amplified_col] > 0]['smiles'].tolist()
decreased_smiles = df[df[max_decreased_col] > 0]['smiles'].tolist()

print("\n** SMILES avec les blocs amplifiés (vert) **")
print("\n".join(amplified_smiles[:5]))  # Affichage limité à 5 résultats

print("\n** SMILES avec les blocs diminués (rouge) **")
print("\n".join(decreased_smiles[:5]))  # Affichage limité à 5 résultats
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ============================
# 1. Chargement des Données
# ============================

# Vérifiez les colonnes disponibles
print("Colonnes disponibles :", df.columns)

# ============================
# 2. Première Étape de Clustering (Clusters Initiaux)
# ============================

# Standardisation des données pour clustering
features = ['pce_pcbm(%)', 'voc_pcbm(V)', 'jsc_pcbm(A.m-2)']  # Colonnes utilisées pour le clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])

# KMeans pour les clusters initiaux
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
df["Cluster"] = kmeans.fit_predict(X_scaled)

# Calcul des moyennes PL par cluster initial
cluster_means = df.groupby(['Cluster', 'sas1(%)']).agg(
    mean_pl=('Gap(eV)', 'mean')
).reset_index()

# ============================
# 3. Sous-Clustering du Cluster 0
# ============================

# Filtrer les données du cluster 0
cluster_0 = df[df["Cluster"] == 0]

# Deuxième clustering sur le cluster 0
kmeans_0 = KMeans(n_clusters=3, random_state=42, n_init='auto')
cluster_0["Subcluster"] = kmeans_0.fit_predict(cluster_0[features])

# Ajouter les sous-clusters au DataFrame principal
df.loc[df["Cluster"] == 0, "Subcluster"] = cluster_0["Subcluster"]

# ============================
# 4. Visualisations
# ============================

# (A) Courbes PL moyennes pour les sous-clusters de Cluster 0
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=cluster_0.groupby(['Subcluster', 'sas1(%)']).agg(mean_pl=('Gap(eV)', 'mean')).reset_index(),
    x="sas1(%)", y="mean_pl", hue="Subcluster", palette="viridis"
)
plt.title("Mean PL Intensity for Subclusters of Cluster 0")
plt.xlabel("sascore")
plt.ylabel("Mean gap energt (eV)")
plt.legend(title="Subcluster")
plt.grid(True)
plt.show()

# (B) Distribution de PCE pour Cluster 0 et ses sous-clusters
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[df['Cluster'] == 0], x="Subcluster", y="pce_pcbm(%)", palette="viridis", showmeans=True)
plt.title("PCE Distribution for Cluster 0 and Subclusters")
plt.xlabel("Subcluster")
plt.ylabel("PCE [%]")
plt.grid(True)
plt.show()

# (C) Distributions des performances pour tous les clusters
performance_metrics = ['pce_pcbm(%)', 'voc_pcbm(V)', 'jsc_pcbm(A.m-2)']  # Remplacez par les colonnes disponibles
plt.figure(figsize=(16, 10))
for i, metric in enumerate(performance_metrics, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x="Cluster", y=metric, palette="viridis", showmeans=True)
    plt.title(f"Distribution of {metric} by Cluster")
    plt.xlabel("Cluster")
    plt.ylabel(metric)
    plt.grid(True)

plt.tight_layout()
plt.show()

````

````python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import rdMolDraw2D
from rdkit.Chem import rdDepictor
import matplotlib.pyplot as plt
from IPython.display import SVG, display
import tempfile

# Define SMARTS patterns for the detected building blocks
amplified_block_pattern = "c1ccccc1N"
decreased_block_pattern = "c1ccccc1Br"

def draw_smiles_with_highlights(smiles_list, highlight_color, block_smarts, title):
    """
    Draws the given list of SMILES with highlighted substructures.
    """
    mols = []
    highlight_atoms = []

    for smiles in smiles_list:
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            rdDepictor.Compute2DCoords(mol)
            pattern = Chem.MolFromSmarts(block_smarts)
            if pattern:
                matches = mol.GetSubstructMatches(pattern)
                highlight_atoms.append([atom_idx for match in matches for atom_idx in match])
            else:
                highlight_atoms.append([])
            mols.append(mol)

    drawer = rdMolDraw2D.MolDraw2DSVG(500, 200 * len(mols))
    options = drawer.drawOptions()
    
    for mol_idx, mol in enumerate(mols):
        if mol:
            drawer.DrawMolecule(
                mol,
                highlightAtoms=highlight_atoms[mol_idx],
                highlightAtomColors={atom_idx: highlight_color for atom_idx in highlight_atoms[mol_idx]}
            )
    
    drawer.FinishDrawing()
    svg_data = drawer.GetDrawingText()
    
    # Save SVG to a temporary file and display it
    with tempfile.NamedTemporaryFile(delete=False, suffix=".svg") as tmpfile:
        tmpfile.write(svg_data.encode("utf-8"))
        tmpfile_path = tmpfile.name
    
    display(SVG(tmpfile_path))

# Draw amplified SMILES with green highlights
draw_smiles_with_highlights(amplified_smiles[:5], (0.0, 1.0, 0.0), amplified_block_pattern, "SMILES with Amplified Blocks (Green)")

# Draw decreased SMILES with red highlights
draw_smiles_with_highlights(decreased_smiles[:5], (1.0, 0.0, 0.0), decreased_block_pattern, "SMILES with Decreased Blocks (Red)")
````
