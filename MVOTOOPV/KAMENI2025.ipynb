```python
import pandas as pd
import os
import time
import datetime
import numpy as np
from pyscf import dft, solvent, lib
from pyscf.tools import molden

def batch_tddft_calculations(df, working_dir_base, basis, solvent_eps=2.3741):
    """Batch TDDFT calculations for a DataFrame containing SMILES"""
    
    results = []
    failed_entries = []
    
    # Create main working directory
    os.makedirs(working_dir_base, exist_ok=True)
    
    for idx, row in df.iterrows():
        try:
            key = f"mol_{idx}"
            smi = row['smiles']
            
            # Create molecule-specific directory
            mol_dir = os.path.join(working_dir_base, key)
            os.makedirs(mol_dir, exist_ok=True)
            
            # Run TDDFT calculations
            df_result, _, _, elapsed, status = tddft_osc_strength_calculations(
                key, smi, mol_dir, basis, solvent_eps
            )
            
            # Add metadata to results
            result_entry = {
                'smiles': smi,
                'calculation_time': elapsed,
                'status': status,
                **df_result['eV'].to_dict()
            }
            results.append(result_entry)
            
        except Exception as e:
            failed_entries.append({
                'smiles': smi,
                'error': str(e),
                'timestamp': datetime.datetime.now()
            })
    
    # Save error log
    if failed_entries:
        pd.DataFrame(failed_entries).to_csv(
            os.path.join(working_dir_base, 'error_log.csv'), index=False
        )
    
    return pd.DataFrame(results)

def tddft_osc_strength_calculations(df1,key,smi,base_xyz_path, working_dir,basis,solvent_eps):
    """Modified version for DataFrame integration"""
    
    # PySCF molecule object with error handling
    try:
        mol_pyscf = pyscf_molecule_object(key, smi,base_xyz_path, working_dir, basis='6-31G(d,p)')
    except Exception as e:
        raise ValueError(f"Molecular construction failed: {str(e)}") from e

    # File paths
    path_results = os.path.join(working_dir, f'Results_{key}.csv')
    path_DFT_chk = os.path.join(working_dir, f'{key}_DFT.chk')
    path_DFT_txt = os.path.join(working_dir, f'{key}_DFT.txt')

    # Configure calculation
    mdft = dft.RKS(mol_pyscf)
    mdft = mdft.ddCOSMO(solvent_eps)
    mdft.xc = 'B3LYP'
    mdft.max_cycle = 512
    mdft.conv_tol = 1e-7
    mdft.grids.level = 0

    # Calculation workflow
    start_time = time.time()
    try:
        if not os.path.exists(path_DFT_chk):
            mdft.chkfile = path_DFT_chk
            mdft.kernel()
            if not mdft.converged:
                mdft = mdft.newton(mdft)
                mdft.kernel()
        else:
            mdft.__dict__.update(lib.chkfile.load(path_DFT_chk, 'scf'))
            mdft.kernel()
        
        mdft.analyze()
        
    except Exception as e:
        raise RuntimeError(f"SCF convergence failed: {str(e)}") from e

    # Process results
    homo_idx = np.where(mdft.mo_occ > 0)[0][-1]
    lumo_idx = np.where(mdft.mo_occ == 0)[0][0]

    energy_data = {
        'Total energy (eV)': mdft.e_tot * au2ev,
        'HOMO (eV)': mdft.mo_energy[homo_idx] * au2ev,
        'LUMO (eV)': mdft.mo_energy[lumo_idx] * au2ev,
        'Gap (eV)': (mdft.mo_energy[lumo_idx] - mdft.mo_energy[homo_idx]) * au2ev
    }

    # Save molecular orbitals
    molden.from_scf(mdft, os.path.join(working_dir, f'{key}.molden'))

    # Create results DataFrame
    df_result = pd.DataFrame([energy_data])
    df_result.to_csv(path_results, index=False)
    
    elapsed = str(datetime.timedelta(seconds=time.time()-start_time))
    
    return df_result, mdft, path_results, elapsed, 'Success'
```

    # Singlets GS DFT calculations

    mdft.xc = 'B3LYP'

    if os.path.exists(path_DFT_chk):
        path_DFT_status = f'{path_DFT_chk} exist'
        start = time.time()

        # Load the DFT results from the checkpoint file
        mdft.__dict__.update(chkfile.load(path_DFT_chk, 'scf'))
        mdft.kernel()

        elapsed_DFT = str(datetime.timedelta(
            seconds = time.time() - start))
    else:
        path_DFT_status = f'{path_DFT_chk} don\'t exist'

        # Save the DFT results to a checkpoint file
        mdft.chkfile = path_DFT_chk

                        index = ['Total energy',

                                 'MO energy of HOMO',
                                 'MO energy of LUMO'])
    return df_DFT,mdft,path_results,elapsed_DFT,path_DFT_status

```python
from pyscf.data import nist
# Conversion factor from atomic units to eV
au2ev = nist.HARTREE2EV

def tddft_osc_strength_calculations(key, xyz_file, working_dir):
    """PySCF TDDFT-TDA calculations for a single molecule"""
    try:
        # Create working directory if it doesn't exist
        os.makedirs(working_dir, exist_ok=True)
        
        # Read XYZ file
        with open(xyz_file, 'r') as f:
            xyz_data = f.readlines()
        
        # Extract atom symbols and coordinates
        atoms = []
        coords = []
        for line in xyz_data[2:]:  # Skip first two lines
            parts = line.split()
            if len(parts) >= 4:
                atoms.append(parts[0])
                coords.append([float(x) for x in parts[1:4]])
        
        # Create PySCF molecule object
        mol_pyscf = gto.Mole()
        mol_pyscf.atom = list(zip(atoms, coords))
        mol_pyscf.basis = '6-31G(2df,p)'
        mol_pyscf.build()
        
        # DFT calculations with solvation
        path_DFT_chk = os.path.join(working_dir, f'{key}_DFT.chk')
        mdft = dft.RKS(mol_pyscf).apply(solvent.ddCOSMO)
        mdft.xc = 'B3LYP'
        mdft.with_solvent.eps = 2.3741  # toluene
        mdft.max_cycle = 512
        mdft.grids.level = 0
        mdft.conv_tol = 1E-7
        
        start = time.time()
        if os.path.exists(path_DFT_chk):
            mdft.__dict__.update(chkfile.load(path_DFT_chk, 'scf'))
        mdft.kernel()
        if not mdft.converged:
            mdft = mdft.newton(mdft)
            mdft.kernel()
        elapsed_DFT = str(datetime.timedelta(seconds=time.time() - start))
        
        # TDDFT calculations
        path_TDDFT_chk = os.path.join(working_dir, f'{key}_TDDFT.chk')
        mtd = tddft.TDA(mdft).apply(solvent.ddCOSMO)
        mtd.with_solvent.eps = 2.3741
        mtd.nstates = 5
        
        start = time.time()
        if os.path.exists(path_TDDFT_chk):
            mtd.__dict__.update(chkfile.load(path_TDDFT_chk, 'tddft'))
        mtd.kernel()
        elapsed_TDDFT = str(datetime.timedelta(seconds=time.time() - start))
        
        # Calculate excited state properties
        fluorescence_energy = mtd.e[0] if len(mtd.e) > 0 else np.nan
        oscillator_strength = mtd.oscillator_strength(gauge='length')[0] if len(mtd.e) > 0 else np.nan
        
        # Triplet calculations
        mtd_triplet = tddft.TDA(mdft).apply(solvent.ddCOSMO)
        mtd_triplet.with_solvent.eps = 2.3741
        mtd_triplet.nstates = 1
        mtd_triplet.singlet = False
        
        try:
            mtd_triplet.kernel()
            triplet_energy = mtd_triplet.e[0] if len(mtd_triplet.e) > 0 else np.nan
            singlet_triplet_gap = fluorescence_energy - triplet_energy
        except:
            triplet_energy = np.nan
            singlet_triplet_gap = np.nan
        
        # Lifetime calculation
        c = nist.LIGHT_SPEED  # in au
        time_au = 2.4188843265857E-17  # in au
        tau_au = c**3 / (2 * (fluorescence_energy)**2 * oscillator_strength) if fluorescence_energy and oscillator_strength else np.nan
        tau = tau_au * time_au if not np.isnan(tau_au) else np.nan
        
        # Multi-Objective function
        mobj = (oscillator_strength - singlet_triplet_gap * au2ev - 
                np.abs(fluorescence_energy * au2ev - 3.2)) if all(not np.isnan(x) for x in [oscillator_strength, singlet_triplet_gap, fluorescence_energy]) else np.nan
        
        return {
            'Molecule': key,
            'Fluorescence_energy_eV': fluorescence_energy * au2ev,
            'Singlet_Triplet_gap_eV': singlet_triplet_gap * au2ev,
            'Oscillator_strength': oscillator_strength,
            'Total_energy_eV': mdft.e_tot * au2ev,
            'Electronic_energy_eV': mdft.energy_elec()[0] * au2ev,
            'Nuclear_repulsion_energy_eV': mdft.energy_nuc() * au2ev,
            'Lifetime_ns': tau * 1e9 if not np.isnan(tau) else np.nan,
            'Multi_Obj': mobj,
            'DFT_calculation_time': elapsed_DFT,
            'TDDFT_calculation_time': elapsed_TDDFT,
            'Status': 'Success'
        }
    except Exception as e:
        return {
            'Molecule': key,
            'Status': f'Failed: {str(e)}'
        }

def calculate_all_molecules(molecule_ids, base_xyz_path, working_dir):
    """Calculate properties for all molecules and return combined DataFrame"""
    all_results = []
    
    for mol_id in molecule_ids:
        key = f"gdb_{mol_id}"
        xyz_file = os.path.join(base_xyz_path, f"GDB{mol_id}.xyz")
        
        if os.path.exists(xyz_file):
            print(f"Processing {key}...")
            result = tddft_osc_strength_calculations(key, xyz_file, working_dir)
            all_results.append(result)
        else:
            print(f"XYZ file not found for {key}")
            all_results.append({
                'Molecule': key,
                'Status': 'Failed: XYZ file not found'
            })
    
    # Create DataFrame from all results
    df_results = pd.DataFrame(all_results)
    
    # Save combined results
    combined_csv_path = os.path.join(working_dir, 'combined_results.csv')
    df_results.to_csv(combined_csv_path, index=False)
    
    return df_results

# Example usage
molecule_ids = my1_df["Id"]



df_all_results = calculate_all_molecules(molecule_ids, base_xyz_path, working_dir)
print(df_all_results)
```

```python
import os
import pandas as pd
import numpy as np



# === 🧪 Calcul de propriétés dérivées
derived_data = []
for _, row in df_raw.iterrows():
    try:
        osc = row.get("Oscillator_Strength", np.nan)
        st_gap = row.get("Singlet_Triplet_gap_eV", np.nan)
        flu = row.get("Fluorescence_energy_eV", np.nan)
        gap = row.get("Gap_HOMO_LUMO_eV", np.nan)
        dip = row.get("Dipole_Debye", np.nan)
        tot_energy = row.get("Total_energy_eV", np.nan)

        # === Propriétés électroniques dérivées
        hardness = gap / 2 if not np.isnan(gap) else np.nan
        mu = -gap / 2 if not np.isnan(gap) else np.nan
        omega = (mu ** 2) / (2 * hardness) if hardness and hardness != 0 else np.nan

        # === Multi-objective scoring
        mobj = (osc - st_gap - np.abs(flu - 3.2)) if all([not np.isnan(x) for x in [osc, st_gap, flu]]) else np.nan

        # === Score de stabilité (exemple)
        stability_score = (-tot_energy + omega) if not np.isnan(tot_energy) and not np.isnan(omega) else np.nan

        derived_data.append({
            "Molecule": row["Molecule"],
            "Gap_HOMO_LUMO_eV": gap,
            "Oscillator_Strength": osc,
            "Fluorescence_energy_eV": flu,
            "Singlet_Triplet_gap_eV": st_gap,
            "Dipole_Debye": dip,
            "Hardness_eta_eV": hardness,
            "Chemical_Potential_mu_eV": mu,
            "Electrophilicity_omega_eV": omega,
            "Multi_Objective_Score": mobj,
            "Stability_Score": stability_score,
            "Lifetime_ns": row.get("Lifetime_ns", np.nan),
            "Total_Energy_eV": tot_energy,
            "Status": row.get("Status", "Unknown")
        })
    except Exception as e:
        print(f"❌ Erreur dans ligne {row['Molecule']}: {str(e)}")
        continue

# === 📊 Résultat final
df_enhanced = pd.DataFrame(derived_data)

# === 💾 Sauvegarde
df_enhanced.to_csv("biocapteurs_properties_enriched.csv", index=False)
df_enhanced.to_pickle("biocapteurs_properties_enriched.pkl")
print("✅ Propriétés enrichies sauvegardées dans : biocapteurs_properties_enriched.csv")

# === Aperçu des premières lignes
print(df_enhanced.head(3))
```

```python
from pyscf import gto
import os

def pyscf_molecule_object(base_xyz_path, basis='6-31G(2df,p)'):
    """
    Crée un objet PySCF Mole à partir d'un fichier .xyz.

    Args:
        base_xyz_path (str): Chemin vers le fichier XYZ
        basis (str): Ensemble de base (default: '6-31G(2df,p)')

    Returns:
        pyscf.gto.Mole: Objet molécule PySCF
    """
    if not os.path.isfile(base_xyz_path):
        raise FileNotFoundError(f"Fichier XYZ introuvable : {base_xyz_path}")

    # Lecture brute pour extraire les atomes et coordonnées
    with open(base_xyz_path, 'r') as f:
        lines = f.readlines()[2:]  # sauter les deux premières lignes
        atoms = []
        coords = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) >= 4:
                atoms.append(parts[0])
                coords.append([float(p) for p in parts[1:4]])

    mol = gto.Mole()
    mol.atom = list(zip(atoms, coords))
    mol.unit = 'Angstrom'
    mol.basis = basis
    mol.symmetry = False
    mol.build()
    return mol
```

```python
import os
import pandas as pd
from pyscf import gto, dft, solvent


# === Paramètres généraux
au2ev = 27.211386245988
xyz_dir = "GDB9_XYZ_2025"
results = []

# === Fonctions utilitaires
def pyscf_molecule_object(base_xyz_path, basis='6-31G(2df,p)'):
    """Construit un objet PySCF Mole à partir d’un fichier .xyz"""
    atoms, coords = [], []
    with open(base_xyz_path, 'r') as f:
        for line in f.readlines()[2:]:  # Ignorer les 2 premières lignes
            parts = line.split()
            if len(parts) >= 4:
                atoms.append(parts[0])
                coords.append([float(x) for x in parts[1:4]])
    mol = gto.Mole()
    mol.atom = list(zip(atoms, coords))
    mol.unit = 'Angstrom'
    mol.basis = basis
    mol.symmetry = False
    mol.build()
    return mol

def calculate_advanced_properties(mdft):
    """Propriétés avancées compatibles PySCF sans AIM ni NMR module"""

    results = {}

    # === 1. Charges (à défaut de AIM)
    try:
        mulliken = mdft.mulliken_pop()[0].tolist()
        lowdin = mdft.lowdin_pop()[0].tolist()
        results['Mulliken_Charges'] = mulliken
        results['Lowdin_Charges'] = lowdin
    except Exception as e:
        results['Mulliken_Charges'] = f"Erreur Mulliken: {str(e)}"
        results['Lowdin_Charges'] = f"Erreur Lowdin: {str(e)}"

    # === 2. Énergie libre de solvatation
    try:
        solv_energy = mdft.with_solvent.e_free
        results['Solvation_Energy_eV'] = solv_energy * au2ev
    except Exception as e:
        results['Solvation_Energy_eV'] = f"Erreur Solvatation: {str(e)}"

    return results


# === Parcours de tous les fichiers XYZ
for fname in os.listdir(xyz_dir):
    if fname.endswith(".xyz"):
        xyz_path = os.path.join(xyz_dir, fname)
        mol_name = os.path.splitext(fname)[0]

        try:
            mol = pyscf_molecule_object(xyz_path)
            mdft = dft.RKS(mol)
            mdft.xc = 'B3LYP'
            mdft.max_cycle = 200
            mdft.conv_tol = 1e-8
            mdft = mdft.apply(solvent.ddCOSMO)
            mdft.with_solvent.eps = 2.3741  # Toluen
            mdft.kernel()

            if not mdft.converged:
                print(f"❌ NON convergé: {mol_name}")
                continue

            props = calculate_advanced_properties(mdft)
            results.append({
                "Molecule": mol_name,
                "Total_Energy_eV": mdft.e_tot * au2ev,
                **props,
                "Status": "Success"
            })

        except Exception as e:
            results.append({
                "Molecule": mol_name,
                "Status": f"Failed: {str(e)}"
            })

# === Résultat final
df_results = pd.DataFrame(results)
df_results.to_csv("advanced_properties_all_molecules.csv", index=False)
df_results.to_pickle("advanced_properties_all_molecules.pkl")

print("✅ Analyse complète terminée. Données enregistrées.")
print(df_results.head())
```

```python
import os
import pandas as pd
from pyscf import gto, dft, solvent

# === Paramètres généraux
au2ev = 27.211386245988
xyz_dir = "GDB9_XYZ_2025"
results = []

# === Chargement du DataFrame contenant les IDs à traiter

id_list = set(my1_df["Id"].astype(str).tolist())  # Convertir les IDs en str pour correspondre aux noms de fichiers

# === Fonctions utilitaires
def pyscf_molecule_object(base_xyz_path, basis='6-31G(2df,p)'):
    atoms, coords = [], []
    with open(base_xyz_path, 'r') as f:
        for line in f.readlines()[2:]:
            parts = line.split()
            if len(parts) >= 4:
                atoms.append(parts[0])
                coords.append([float(x) for x in parts[1:4]])
    mol = gto.Mole()
    mol.atom = list(zip(atoms, coords))
    mol.unit = 'Angstrom'
    mol.basis = basis
    mol.symmetry = False
    mol.build()
    return mol

def calculate_advanced_properties(mdft):
    """Propriétés avancées compatibles avec ddCOSMO."""
    results = {}

    # === 1. Charges Mulliken et Lowdin via mdft.base (RKS original)
    try:
        mulliken = mdft.base.mulliken_pop()[0].tolist()
        lowdin = mdft.base.lowdin_pop()[0].tolist()
        results['Mulliken_Charges'] = mulliken
        results['Lowdin_Charges'] = lowdin
    except Exception as e:
        results['Mulliken_Charges'] = f"Erreur Mulliken: {str(e)}"
        results['Lowdin_Charges'] = f"Erreur Lowdin: {str(e)}"

    # === 2. Énergie libre de solvatation (accessible via .e_free_energy)
    try:
        solv_energy = mdft.e_free_energy
        results['Solvation_Energy_eV'] = solv_energy * au2ev
    except Exception as e:
        results['Solvation_Energy_eV'] = f"Erreur Solvatation: {str(e)}"

    return results

# === Traitement conditionnel des fichiers GDB{Id}.xyz
for fname in os.listdir(xyz_dir):
    if fname.endswith(".xyz") and fname.startswith("GDB"):
        mol_id = fname.replace("GDB", "").replace(".xyz", "")
        if mol_id not in id_list:
            continue  # Ignorer les fichiers non listés dans le DataFrame

        xyz_path = os.path.join(xyz_dir, fname)
        mol_name = os.path.splitext(fname)[0]

        try:
            mol = pyscf_molecule_object(xyz_path)
            mdft = dft.RKS(mol)
            mdft.xc = 'B3LYP'
            mdft.max_cycle = 200
            mdft.conv_tol = 1e-8
            mdft = mdft.apply(solvent.ddCOSMO)
            mdft.with_solvent.eps = 2.3741  # Toluen
            mdft.kernel()

            if not mdft.converged:
                print(f"❌ NON convergé: {mol_name}")
                continue

            props = calculate_advanced_properties(mdft)
            results.append({
                "Id": mol_id,
                "Molecule": mol_name,
                "Total_Energy_eV": mdft.e_tot * au2ev,
                **props,
                "Status": "Success"
            })

        except Exception as e:
            results.append({
                "Id": mol_id,
                "Molecule": mol_name,
                "Status": f"Failed: {str(e)}"
            })

# === Résultat final
df_results = pd.DataFrame(results)
df_results.to_csv("advanced_properties_selected_ids.csv", index=False)
df_results.to_pickle("advanced_properties_selected_ids.pkl")

print("✅ Analyse terminée pour les molécules sélectionnées.")
print(df_results.head())
```
