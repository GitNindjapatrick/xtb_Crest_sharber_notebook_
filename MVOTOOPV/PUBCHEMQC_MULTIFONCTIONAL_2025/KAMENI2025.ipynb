```python
def pyscf_molecule_object(base_xyz_path, working_dir, basis='6-31G*'):
    """
    Crée un objet molécule PySCF à partir d'un smiS.
    
    Args:
        smi_key (str): Clé smiS
        smi (str): Chaîne smiS
        working_dir (str): Répertoire de travail
        basis (str, optional): Ensemble de base à utiliser, par défaut '631g*'
    
    Returns:
        pyscf.gto.Mole: Objet molécule PySCF
    """
    os.makedirs(working_dir, exist_ok=True)
        
    # Read XYZ file
    with open(base_xyz_path, 'r') as f:
        xyz_data = f.readlines()
        
        # Extract atom symbols and coordinates
    atoms = []
    coords = []
    for line in xyz_data[2:]:  # Skip first two lines
        parts = line.split()
        if len(parts) >= 4:
            atoms.append(parts[0])
            coords.append([float(x) for x in parts[1:4]])
                                                                                                                                                                                                         
    mol_pyscf = gto.Mole()
    mol_pyscf.build(
        atom=base_xyz_path,
        basis=basis,
        symmetry=True,
        unit='Angstrom'
    )
    
    return mol_pyscf

```

def tddft_osc_strength_calculations(key, smi, working_dir):
    """PySCF TDDFT-TDA calculations of fluorescence energy,
    singlet-triplet gap and oscillator strength

    Part of code is from https://github.com/aspuru-guzik-group/Tartarus

    Args:
        key (str): key
        smi (str): The smi string of the molecule
        working_dir (str): The directory path where results of calculations will be saved
        phase (str) : 'gas' for the gas-phase or 'sol' for solvation

    """

     # PySCF molecule object
    mol_pyscf = pyscf_molecule_object(key, smi, working_dir, basis = '6-31G(d,p)')

    #=====================================
    # DFT CALCULATIONS
    #=====================================

   # Setting path for DFT results, chkfile and analyze object
    path_results = os.path.join(working_dir, f'Results_{key}.csv')
    path_DFT_chk = os.path.join(working_dir, f'{key}_DFT.chk')
    path_DFT_txt = os.path.join(working_dir, f'{key}_DFT.txt')

    mol_pyscf.output = path_DFT_txt

    # Singlets GS DFT calculations
    mdft = dft.RKS(mol_pyscf).apply(solvent.ddCOSMO)
    mdft.xc = 'B3LYP'
    mdft.with_solvent.eps = 2.3741  # toluene
    mdft.max_cycle = 512
    mdft.grids.level = 0
    mdft.conv_tol = 1E-7
    # if phase == 'sol':
    #     mdft = mdft.ddCOSMO()
    #     mdft.with_solvent.eps = 2.3741   # Toluene
    # mdft.density_fit(auxbasis='def2-universal-jkfit')

    if os.path.exists(path_DFT_chk):
        path_DFT_status = f'{path_DFT_chk} exist'
        start = time.time()

        # Load the DFT results from the checkpoint file
        mdft.__dict__.update(chkfile.load(path_DFT_chk, 'scf'))
        mdft.kernel()

        elapsed_DFT = str(datetime.timedelta(
            seconds = time.time() - start))
    else:
        path_DFT_status = f'{path_DFT_chk} don\'t exist'
        start = time.time()

        # Save the DFT results to a checkpoint file
        mdft.chkfile = path_DFT_chk
        mdft.kernel()
        if(not mdft.converged):
            mdft = mdft.newton(mdft)
            mdft.kernel()

        elapsed_DFT = str(datetime.timedelta(
            seconds = time.time()- start))

    # Analyze the given DFT object
    mdft.analyze()

    # Index of HOMO and LUMO
    homo_idx = np.where(mdft.mo_occ > 0)[0][-1]
    lumo_idx = np.where(mdft.mo_occ == 0)[0][0]

    # Create the DFT results dataframe
    list_results_DFT = [mdft.e_tot,
                        mdft.energy_elec()[0],
                        mdft.energy_nuc(),
                        mdft.mo_energy[homo_idx],
                        mdft.mo_energy[lumo_idx]]
    dict_results_DFT = {'Hartree': list_results_DFT,
                    'eV': [x * au2ev for x in list_results_DFT]}
    df_DFT = pd.DataFrame(dict_results_DFT,
                        index = ['Total energy',
                                 'Electronic energy',
                                 'Nuclear repulsion energy',
                                 'MO energy of HOMO',
                                 'MO energy of LUMO'])
    return df_DFT,mdft,path_results,elapsed_DFT,path_DFT_status

```python
from pyscf import gto
import os

def pyscf_molecule_object(base_xyz_path, basis='6-31G*'):
    """
    Crée un objet PySCF Mole à partir d'un fichier .xyz.

    Args:
        base_xyz_path (str): Chemin vers le fichier XYZ
        basis (str): Ensemble de base (default: '6-31G*')

    Returns:
        pyscf.gto.Mole: Objet molécule PySCF
    """
    if not os.path.isfile(base_xyz_path):
        raise FileNotFoundError(f"Fichier XYZ introuvable : {base_xyz_path}")

    # Lecture brute pour extraire les atomes et coordonnées
    with open(base_xyz_path, 'r') as f:
        lines = f.readlines()[2:]  # sauter les deux premières lignes
        atoms = []
        coords = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) >= 4:
                atoms.append(parts[0])
                coords.append([float(p) for p in parts[1:4]])

    mol = gto.Mole()
    mol.atom = list(zip(atoms, coords))
    mol.unit = 'Angstrom'
    mol.basis = basis
    mol.symmetry = False
    mol.build()
    return mol
```

def tddft_osc_strength_calculations(key, xyz_file, working_dir):
    """PySCF TDDFT-TDA calculations for a single molecule"""
    try:
        # Create working directory if it doesn't exist
        os.makedirs(working_dir, exist_ok=True)
        
        # Read XYZ file
        with open(xyz_file, 'r') as f:
            xyz_data = f.readlines()
        
        # Extract atom symbols and coordinates
        atoms = []
        coords = []
        for line in xyz_data[2:]:  # Skip first two lines
            parts = line.split()
            if len(parts) >= 4:
                atoms.append(parts[0])
                coords.append([float(x) for x in parts[1:4]])
        
        # Create PySCF molecule object
        mol_pyscf = gto.Mole()
        mol_pyscf.atom = list(zip(atoms, coords))
        mol_pyscf.basis = '6-31G*'
        mol_pyscf.build()
        
        # DFT calculations with solvation
        path_DFT_chk = os.path.join(working_dir, f'{key}_DFT.chk')
        mdft = dft.RKS(mol_pyscf).apply(solvent.ddCOSMO)
        mdft.xc = 'B3LYP'
        mdft.with_solvent.eps = 2.3741  # toluene
        mdft.max_cycle = 512
        mdft.grids.level = 0
        mdft.conv_tol = 1E-7
        
        start = time.time()
        if os.path.exists(path_DFT_chk):
            mdft.__dict__.update(chkfile.load(path_DFT_chk, 'scf'))
        mdft.kernel()
        if not mdft.converged:
            mdft = mdft.newton(mdft)
            mdft.kernel()
        elapsed_DFT = str(datetime.timedelta(seconds=time.time() - start))
        
        # TDDFT calculations
        path_TDDFT_chk = os.path.join(working_dir, f'{key}_TDDFT.chk')
        mtd = tddft.TDA(mdft).apply(solvent.ddCOSMO)
        mtd.with_solvent.eps = 2.3741
        mtd.nstates = 5
        
        start = time.time()
        if os.path.exists(path_TDDFT_chk):
            mtd.__dict__.update(chkfile.load(path_TDDFT_chk, 'tddft'))
        mtd.kernel()
        elapsed_TDDFT = str(datetime.timedelta(seconds=time.time() - start))
        
        # Calculate excited state properties
        fluorescence_energy = mtd.e[0] if len(mtd.e) > 0 else np.nan
        oscillator_strength = mtd.oscillator_strength(gauge='length')[0] if len(mtd.e) > 0 else np.nan
        
        # Triplet calculations
        mtd_triplet = tddft.TDA(mdft).apply(solvent.ddCOSMO)
        mtd_triplet.with_solvent.eps = 2.3741
        mtd_triplet.nstates = 1
        mtd_triplet.singlet = False
        
        try:
            mtd_triplet.kernel()
            triplet_energy = mtd_triplet.e[0] if len(mtd_triplet.e) > 0 else np.nan
            singlet_triplet_gap = fluorescence_energy - triplet_energy
        except:
            triplet_energy = np.nan
            singlet_triplet_gap = np.nan
        
        # Lifetime calculation
        c = nist.LIGHT_SPEED  # in au
        time_au = 2.4188843265857E-17  # in au
        tau_au = c**3 / (2 * (fluorescence_energy)**2 * oscillator_strength) if fluorescence_energy and oscillator_strength else np.nan
        tau = tau_au * time_au if not np.isnan(tau_au) else np.nan
        
        # Multi-Objective function
        mobj = (oscillator_strength - singlet_triplet_gap * au2ev - 
                np.abs(fluorescence_energy * au2ev - 3.2)) if all(not np.isnan(x) for x in [oscillator_strength, singlet_triplet_gap, fluorescence_energy]) else np.nan
        
        return {
            'Molecule': key,
            'Fluorescence_energy_eV': fluorescence_energy * au2ev,
            'Singlet_Triplet_gap_eV': singlet_triplet_gap * au2ev,
            'Oscillator_strength': oscillator_strength,
            'Total_energy_eV': mdft.e_tot * au2ev,
            'Electronic_energy_eV': mdft.energy_elec()[0] * au2ev,
            'Nuclear_repulsion_energy_eV': mdft.energy_nuc() * au2ev,
            'Lifetime_ns': tau * 1e9 if not np.isnan(tau) else np.nan,
            'Multi_Obj': mobj,
            'DFT_calculation_time': elapsed_DFT,
            'TDDFT_calculation_time': elapsed_TDDFT,
            'Status': 'Success'
        }
    except Exception as e:
        return {
            'Molecule': key,
            'Status': f'Failed: {str(e)}'
        }

```python
import os
import numpy as np
import pandas as pd
import time, datetime
from pyscf import gto, dft, tddft, solvent
from pyscf.data import nist

au2ev = nist.HARTREE2EV
LIGHT_SPEED = nist.LIGHT_SPEED
TIME_AU = 2.4188843265857E-17  # au to seconds

def tddft_osc_strength_calculations(key, xyz_file, working_dir, log_file="error_log.txt"):
    try:
        os.makedirs(working_dir, exist_ok=True)
        
        with open(xyz_file, 'r') as f:
            lines = f.readlines()
        atoms = [line.split()[0] for line in lines[2:]]
        coords = [[float(x) for x in line.split()[1:4]] for line in lines[2:]]
        
        mol = gto.Mole()
        mol.atom = list(zip(atoms, coords))
        mol.basis = '6-31G*'
        mol.build()
        
        path_dft_chk = os.path.join(working_dir, f"{key}_DFT.chk")
        mf = dft.RKS(mol).apply(solvent.ddCOSMO)
        mf.xc = 'B3LYP'
        mf.with_solvent.eps = 2.3741
        mf.max_cycle = 512
        mf.grids.level = 0
        mf.conv_tol = 1E-7
        
        start = time.time()
        if os.path.exists(path_dft_chk):
            mf.__dict__.update(chkfile.load(path_dft_chk, 'scf'))
        mf.kernel()
        if not mf.converged:
            mf = mf.newton(mf)
            mf.kernel()
        elapsed_dft = str(datetime.timedelta(seconds=time.time() - start))
        
        mtd = tddft.TDA(mf).apply(solvent.ddCOSMO)
        mtd.with_solvent.eps = 2.3741
        mtd.nstates = 5
        
        start = time.time()
        mtd.kernel()
        elapsed_tddft = str(datetime.timedelta(seconds=time.time() - start))
        
        e1 = mtd.e[0] if len(mtd.e) > 0 else np.nan
        osc = mtd.oscillator_strength(gauge='length')[0] if len(mtd.e) > 0 else np.nan

        mtd_triplet = tddft.TDA(mf).apply(solvent.ddCOSMO)
        mtd_triplet.with_solvent.eps = 2.3741
        mtd_triplet.nstates = 1
        mtd_triplet.singlet = False
        try:
            mtd_triplet.kernel()
            triplet = mtd_triplet.e[0] if len(mtd_triplet.e) > 0 else np.nan
            gap = e1 - triplet
        except:
            triplet, gap = np.nan, np.nan

        tau_au = LIGHT_SPEED**3 / (2 * (e1)**2 * osc) if e1 and osc else np.nan
        tau_ns = tau_au * TIME_AU * 1e9 if not np.isnan(tau_au) else np.nan

        mobj = (osc - gap * au2ev - abs(e1 * au2ev - 3.2)) if all(not np.isnan(x) for x in [osc, gap, e1]) else np.nan

        return {
            'Molecule': key,
            'Fluorescence_energy_eV': e1 * au2ev,
            'Singlet_Triplet_gap_eV': gap * au2ev,
            'Oscillator_strength': osc,
            'Total_energy_eV': mf.e_tot * au2ev,
            'Electronic_energy_eV': mf.energy_elec()[0] * au2ev,
            'Nuclear_repulsion_energy_eV': mf.energy_nuc() * au2ev,
            'Lifetime_ns': tau_ns,
            'Multi_Obj': mobj,
            'DFT_calculation_time': elapsed_dft,
            'TDDFT_calculation_time': elapsed_tddft,
            'Status': 'Success'
        }

    except Exception as e:
        with open(os.path.join(working_dir, log_file), 'a') as logf:
            logf.write(f"{key} FAILED: {str(e)}\n")
        return {
            'Molecule': key,
            'Fluorescence_energy_eV': np.nan,
            'Singlet_Triplet_gap_eV': np.nan,
            'Oscillator_strength': np.nan,
            'Total_energy_eV': np.nan,
            'Electronic_energy_eV': np.nan,
            'Nuclear_repulsion_energy_eV': np.nan,
            'Lifetime_ns': np.nan,
            'Multi_Obj': np.nan,
            'DFT_calculation_time': None,
            'TDDFT_calculation_time': None,
            'Status': f"Failed: {str(e)}"
        }
```

```python
def calculate_all_molecules(molecule_ids, base_xyz_path, working_dir):
    all_results = []
    os.makedirs(working_dir, exist_ok=True)
    
    for key in molecule_ids:
        xyz_file = os.path.join(base_xyz_path, f"{key}.xyz")
        
        if os.path.exists(xyz_file):
            print(f"🧪 Calcul de la molécule {key}")
            result = tddft_osc_strength_calculations(key, xyz_file, working_dir)
        else:
            print(f"⚠️ Fichier XYZ non trouvé pour {key}")
            result = {
                'Molecule': key,
                'Fluorescence_energy_eV': np.nan,
                'Singlet_Triplet_gap_eV': np.nan,
                'Oscillator_strength': np.nan,
                'Total_energy_eV': np.nan,
                'Electronic_energy_eV': np.nan,
                'Nuclear_repulsion_energy_eV': np.nan,
                'Lifetime_ns': np.nan,
                'Multi_Obj': np.nan,
                'DFT_calculation_time': None,
                'TDDFT_calculation_time': None,
                'Status': "Failed: XYZ file not found"
            }

        all_results.append(result)

    df_results = pd.DataFrame(all_results)
    df_results.to_csv(os.path.join(working_dir, "combined_results.csv"), index=False)
    print(f"✅ Résultats sauvegardés dans : {working_dir}/combined_results.csv")
    return df_results
```

# === Compute multi-objective score
def compute_multiobjective(row):
    osc = row.get("Oscillator_strength")
    gap = row.get("Singlet_Triplet_gap_eV")
    e1 = row.get("Fluorescence_energy_eV")
    if all(not pd.isna(x) for x in [osc, gap, e1]):
        return osc - gap * au2ev - abs(e1 * au2ev - 3.2)
    return np.nan

df["MultiObjectiveScore"] = df.apply(compute_multiobjective, axis=1)

# === Domain classification
def classify_small_molecule_domains(row):
    domains = []

    if 0.4 < row['Oscillator_strength'] <= 1.5 and 2.0 < row['Fluorescence_energy_eV'] < 3.1 and row['Singlet_Triplet_gap_eV'] < 0.6:
        domains.append("OLED")

    if row['Lifetime_ns'] < 5 and row['Fluorescence_energy_eV'] > 2.0:
        domains.append("Sensor")

    if row['Lifetime_ns'] > 6 and row['Oscillator_strength'] > 0.3:
        domains.append("Bioimaging")

    if row['Singlet_Triplet_gap_eV'] > 0.2 and row['Fluorescence_energy_eV'] > 2.0:
        domains.append("Photocatalysis")

    if row['Oscillator_strength'] > 0.5 and 1.8 < row['Fluorescence_energy_eV'] < 3.3:
        domains.append("Dye")

    return ", ".join(domains) if domains else None

df["Application_Domain"] = df.apply(classify_small_molecule_domains, axis=1)

```python
# === Attribution des domaines d'application
def assign_domains(row):
    domains = []
    if row.get("PCE_PCBM", 0) - row.get("SAscore", 10) > 2.5:
        domains.append("OPV-PCBM")
    if row.get("PCE_PCDTBT", 0) - row.get("SAscore", 10) > 2.5:
        domains.append("OPV-PCDTBT")
    if row.get("Multi_Obj", -10) > -0.6:
        domains.append("OLED")
    if row.get("Lifetime_ns", 0) > 6 and row.get("Oscillator_strength", 0) > 0.006:
        domains.append("Bioimaging")
    if row.get("Lifetime_ns", 100) < 4 and row.get("Fluorescence_energy_eV", 0) > 2:
        domains.append("Sensor")
    if row.get("Singlet_Triplet_gap_eV", 0) > 0.3 and row.get("Fluorescence_energy_eV", 0) > 2.6:
        domains.append("Photocatalysis")
    return domains if domains else np.nan  # ⬅️ retourne une liste, pas une string

df["Domains"] = df.apply(assign_domains, axis=1)
df = df.dropna(subset=["Domains"])
df_exploded = df.explode("Domains")  # Pas besoin de str.split ici
```
